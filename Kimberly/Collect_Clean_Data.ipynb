```python
import pandas as pd
import numpy as np
```

```python
flight_data_df = pd.read_csv('../flight_data.csv.zip')
flight_data_df.head()
```

```python
# flight_data_df.to_csv('../flight_data.csv.zip', index=False)
```

```python
flight_data_df.shape
```

```python
#rename the columns in our dataframe
flight_data_df.rename(columns={'from_airport_code': 'Departing Airport', 
                   'from_country': 'Departing Country',
                   'dest_airport_code': 'Destination Airport',
                   'dest_country':'Destination Country',
                   'airline_name': 'Airline Name',
                   'flight_number': 'Flight Number',
                   'departure_time': 'Departure Time',
                   'arrival_time': 'Arrival Time',
                               
                  }, inplace=True)
```

```python
flight_data_df.drop(['airline_number','co2_emissions', 'avg_co2_emission_for_this_route','co2_percentage','scan_date' ], axis=1, inplace=True)
```

```python
flight_data_df.dtypes
```

```python
flight_data_df.isnull().sum()
```

```python
cleaned_df = flight_data_df.dropna()
cleaned_df.head()
```

```python
cleaned_df.isnull().sum()
```

```python
try:
    cleaned_df["price"] = cleaned_df["price"].map("${:,.2f}".format)
except ValueError:
    pass
```

```python
cleaned_df['Depart Time'] = pd.to_datetime(cleaned_df['Departure Time']).dt.strftime('%p')
cleaned_df['Arrive Time'] = pd.to_datetime(cleaned_df['Arrival Time']).dt.strftime('%p')
cleaned_df.head(100)
```
