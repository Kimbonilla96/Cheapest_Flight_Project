```python
#rename the columns in our dataframe
flight_data_df.rename(columns={'from_airport_code': 'Departing Airport', 
                   'from_country': 'Departing Country',
                   'dest_airport_code': 'Destination Airport',
                   'dest_country':'Destination Country',
                   'airline_name': 'Airline Name',
                   'flight_number': 'Flight Number',
                   'departure_time': 'Departure Time',
                   'arrival_time': 'Arrival Time',
                               
                  }, inplace=True)
```

```python
flight_data_df.dtypes
```

```python
flight_data_df.isnull().sum()
```

```python
cleaned_df = flight_data_df.dropna()
cleaned_df.head()
```

```python
cleaned_df.isnull().sum()
```

```python
try:
    cleaned_df["price"] = cleaned_df["price"].map("${:,.2f}".format)
except ValueError:
    pass
```

```python
cleaned_df['Depart Time'] = pd.to_datetime(cleaned_df['Departure Time']).dt.strftime('%p')
cleaned_df['Arrive Time'] = pd.to_datetime(cleaned_df['Arrival Time']).dt.strftime('%p')
cleaned_df.head(100)
```

```python

#cleaned_df
cleaned_df['Departure Time'] = pd.to_datetime(cleaned_df['Departure Time'])

cleaned_df['Departure Day'] = cleaned_df['Departure Time'].dt.day_name()
cleaned_df
```

```python
#cleaned_df
cleaned_df['Arrival Time'] = pd.to_datetime(cleaned_df['Arrival Time'])

cleaned_df['Arrival Day'] = cleaned_df['Arrival Time'].dt.day_name()
cleaned_df
#week_df = cleaned_df[(cleaned_df['Departure Day']=='Monday')]
#week_df.head(200)
```

```python
def convert_to_hours_minutes(duration):
    if len(str(duration)) == 4:
        hours = str(duration)[:-2]
        minutes = str(duration)[-2:]
        return f"{hours}:{minutes}"
    else:
        return duration

# Apply the function to the 'duration' column
cleaned_df['Trip Duration'] = cleaned_df['duration'].apply(convert_to_hours_minutes)

cleaned_df.head()
```

```python
Finished_df = cleaned_df.drop(['duration', 'Departure Time', 'Arrival Time', 'Flight Number'], axis=1)
Finished_df.head()
```

```python
desired_column_order = ['Airline Name','Departing Airport', 'Departing Country', 'Depart Time', 'Destination Airport', 'Destination Country','Arrive Time','Departure Day','Arrival Day','stops','Trip Duration', 'price', 'currency','aircraft_type']
Finished_df=Finished_df[desired_column_order]
Finished_df.head()
```
